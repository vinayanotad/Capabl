```python
import datetime

# Initialize empty inventory
inventory = {}

def add_product():
    product_id = input("Enter Product ID: ")
    product_name = input("Enter Product Name: ")
    product_category = input("Enter Product Category: ")
    quantity = int(input("Enter Quantity in Stock: "))
    price_per_unit = float(input("Enter Price per Unit: "))
    expiry_date = input("Enter Expiry Date (YYYY-MM-DD, leave blank if not applicable): ")

    if not expiry_date:
        product_details = {
            "Product Name": product_name,
            "Product Category": product_category,
            "Quantity in Stock": quantity,
            "Price per Unit": price_per_unit
        }
    else:
        product_details = {
            "Product Name": product_name,
            "Product Category": product_category,
            "Quantity in Stock": quantity,
            "Price per Unit": price_per_unit,
            "Expiry Date": expiry_date
        }

    inventory[product_id] = product_details
    print("Product added successfully!")

def display_inventory():
    if not inventory:
        print("Inventory is empty.")
    else:
        print("\nCurrent Inventory:")
        for product_id, details in inventory.items():
            print(f"Product ID: {product_id}")
            for key, value in details.items():
                print(f"{key}: {value}")
            print("-" * 20)

def search_product():
    search_term = input("Enter Product ID or Product Name to search: ")
    for product_id, details in inventory.items():
        if search_term.lower() in (product_id.lower(), details["Product Name"].lower()):
            print("Product found! Details:")
            print(f"Product ID: {product_id}")
            for key, value in details.items():
                print(f"{key}: {value}")
            return
    print("Product not found.")

def update_product():
    product_id = input("Enter Product ID to update: ")
    if product_id in inventory:
        print("Select the information to update:")
        print("1. Quantity in Stock")
        print("2. Price per Unit")
        choice = input("Enter your choice: ")

        if choice == "1":
            new_quantity = int(input("Enter new Quantity in Stock: "))
            inventory[product_id]["Quantity in Stock"] = new_quantity
        elif choice == "2":
            new_price = float(input("Enter new Price per Unit: "))
            inventory[product_id]["Price per Unit"] = new_price
        else:
            print("Invalid choice. No updates made.")
            return

        print("Product details updated successfully!")
    else:
        print("Product not found.")

def delete_product():
    product_id = input("Enter Product ID to delete: ")
    if product_id in inventory:
        del inventory[product_id]
        print("Product deleted successfully!")
    else:
        print("Product not found.")

def calculate_total_value():
    total_value = 0
    for details in inventory.values():
        total_value += details["Quantity in Stock"] * details["Price per Unit"]
    print(f"Total Inventory Value: ${total_value:.2f}")

# Menu System
while True:
    print("\n==== Inventory Management System ====")
    print("1. Add Product")
    print("2. Display Inventory")
    print("3. Search for a Product")
    print("4. Update Product Details")
    print("5. Delete a Product")
    print("6. Calculate Total Value")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    if choice == "1":
        add_product()
    elif choice == "2":
        display_inventory()
    elif choice == "3":
        search_product()
    elif choice == "4":
        update_product()
    elif choice == "5":
        delete_product()
    elif choice == "6":
        calculate_total_value()
    elif choice == "7":
        print("Exiting Inventory Management System. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number from 1 to 7.")
```
